```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the data
df = pd.read_csv('../dataset/TARP.csv')
df.head(1000)
```

```python
#df = df.dropna()
df = df.drop(columns=["Time", "Temperature", " Soil Humidity", "Pressure (KPa)", "ph", "rainfall", "N", "P", "K"])
df.columns

df.fillna(0, inplace=True)
```

```python
# Multiply 1.2 to every moisture value to estimate when the forrest is wet and the fungus is growing
df["Soil Moisture"] = df["Soil Moisture"] * 1.2 # As long it is not over 100%

df.loc[df["Soil Moisture"] > 100, "Soil Moisture"] = 100

# Multiply 1.3 to wind speed to estimate when the forrest might spread the fungus
df["Wind speed (Km/h)"] = df["Wind speed (Km/h)"] * 1.3
df["Wind gust (Km/h)"] = df["Wind gust (Km/h)"] * 1.3
```

```python
# We give the binary values to a float after the number of on's in a row to make a percentage

df.insert(column="Percentage", value=np.zeros(len(df)), loc=1)
# How many on's are in the last 50 columns
for index in range(80,len(df)):
    numberOfOnsInAInterval = 0
    if index <= 80:
        start = 0
    else:
        start = index - 80
    ons = df["Status"].iloc[start:index]

    interval = index - start
    for on in ons:
        if on == "ON":
            numberOfOnsInAInterval += 1
    percentage = numberOfOnsInAInterval / interval
    df["Percentage"].iloc[index] = percentage * 0.85
```

```python
# Remove the first 50 rows
df = df.iloc[50:].reset_index(drop=True)

# Remove the status column
df = df.drop(columns=["Status"])
df
```

```python
# Save the CSV
df.to_csv("../dataset/TARP_preprocessed.csv", index=False)
```
